```python
# Exercise 2: Classification in numpy

# We provide two 2D toy data samples for this exercise. One of them is a linear pattern, another one is a parabolic pattern.

# (a) Data Preprocessing

# 1) Load data sets from the 'linear.csv' and 'parabolic.csv', extract both labels vector and instances matrix.
import matplotlib.cbook as cbook
import os
# cwd = os.getcwd()
linearDataSet = np.loadtxt('linear.csv',delimiter=",")
parabolicDataSet = np.loadtxt('parabolic.csv',delimiter=",")

# print len(linearDataSet)

linearDataSetPositive = linearDataSet[linearDataSet[:,2] == 1.0, 0:2]
linearDataSetNegative = linearDataSet[linearDataSet[:,2] == -1.0, 0:2]
# print len(linearDataSetPositive)
# print len(linearDataSetNegative)

parabolicDataSetPositive = parabolicDataSet[parabolicDataSet[:,2] == 1.0, 0:2]
parabolicDataSetNegative = parabolicDataSet[parabolicDataSet[:,2] == -1.0, 0:2]
# print len(parabolicDataSet)
# print len(parabolicDataSetPositive)
# print len(parabolicDataSetNegative)


# 2) Now try to plot original data in two subplots, positive sample in red and negative sample in blue. hint: use plot, subplot
plt.figure(1)
plt.subplot(211)
plt.title('Linear Positive')
plt.plot(linearDataSetPositive[0],linearDataSetPositive[1],'r')
plt.subplot(223)
plt.title('Linear Negative')
plt.plot(linearDataSetNegative[0],linearDataSetNegative[1],'b')
plt.show()

plt.figure(1)
plt.subplot(211)
plt.title('Parabolic Positive')
plt.plot(parabolicDataSetPositive[0],parabolicDataSetPositive[1],'r')
plt.subplot(223)
plt.title('Parabolic Negative')
plt.plot(parabolicDataSetNegative[0],parabolicDataSetNegative[1],'b')
plt.show()


# 3) Scale the data in [􀀀1; 1].

# The simplest method is rescaling the range of features to scale the range in [0, 1] or [−1, 1]. 
# Selecting the target range depends on the nature of the data. The general formula is given as:
#          x' = \frac{x - \text{min}(x)}{\text{max}(x)-\text{min}(x)}
# where x is an original value, x' is the normalized value. 
scaledLinearDataSet = (linearDataSet - np.amin(linearDataSet,axis = 0))/(np.amax(linearDataSet,axis = 0) - np.amin(linearDataSet,axis = 0))
# print scaledLinearDataSet
scaledParabolicDataSet = (parabolicDataSet - np.amin(parabolicDataSet,axis = 0))/(np.amax(parabolicDataSet,axis = 0) - np.amin(linearDataSet,axis = 0))
# print parabolicDataSet



# 4) Now write back the scaled data sets to an output CSV file.
np.savetxt("scaledLinearDataSet.csv", scaledLinearDataSet, delimiter=",")
np.savetxt("scaledParabolicDataSet.csv", scaledParabolicDataSet, delimiter=",")


# 5) Try to fit both datasets using logistic regression. What can you tell about the results?

def plotDecisionBoundary(clf,title,XTrain,YTrain):
    import seaborn as sns
    sns.set(style="white")
    xx, yy = np.mgrid[-1:1:.01, -1:1:.01]
    grid = np.c_[xx.ravel(), yy.ravel()]
    probs = clf.predict_proba(grid)[:, 1].reshape(xx.shape)
    f, ax = plt.subplots(figsize=(8, 6))
    ax.contour(xx, yy, probs, levels=[.5], cmap="Greys", vmin=0, vmax=.6)

    ax.scatter(XTrain[:,0], XTrain[:,1], c=YTrain[:], s=50,
               cmap="RdBu", vmin=-.2, vmax=1.2,
               edgecolor="white", linewidth=1)

    ax.set(aspect="equal",
           xlim=(0, 1), ylim=(0, 1),
           xlabel="$X_1$", ylabel="$X_2$",title=title)

from sklearn import linear_model
LogRegLinear = linear_model.LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, random_state=None)
XTrainLinear = scaledLinearDataSet[:,0:2]
YLinear = scaledLinearDataSet[:,2]
# print XTrainLinear.shape, YLinear.shape
clf = LogRegLinear.fit(XTrainLinear, YLinear)
plotDecisionBoundary(clf,'Logistic Regression on Linear Data Set',XTrainLinear, YLinear)

XTrainParabola = scaledParabolicDataSet[:,0:2]
YParabola = scaledParabolicDataSet[:,2]
clf1 = LogRegLinear.fit(XTrainParabola, YParabola)
plotDecisionBoundary(clf1,'Logistic Regression on Parabolic Data Set',XTrainParabola, YParabola)

# 6) Now try to do the same with SVM. Which kernel works well on the first dataset? What about the second? Why?
from sklearn import svm
linear_svc = svm.SVC(kernel='linear',probability = True)
# linear_svc.kernel
svcL = linear_svc.fit(XTrainLinear, YLinear)
plotDecisionBoundary(svcL,'SVM using Linear Kernel on Linear Data Set',XTrainLinear, YLinear)
                        
svcL1 = linear_svc.fit(XTrainParabola, YParabola)
plotDecisionBoundary(svcL1,'SVM using Linear Kernel on Parabolic Data Set',XTrainParabola, YParabola)
                        
nonSVC = svm.NuSVC(probability = True)
clfnl = nonSVC.fit(XTrainLinear, YLinear)
plotDecisionBoundary(clfnl,'SVM using Non Linear Kernel on Linear Data Set',XTrainLinear, YLinear)

clfnl1 = nonSVC.fit(XTrainParabola, YParabola)
plotDecisionBoundary(clfnl1,'SVM using Non Linear Kernel on Parabolic Data Set',XTrainParabola, YParabola)

# rbf_svc = svm.SVC(kernel='rbf',probability = True)
# # rbf_svc.kernel
# svcR = rbf_svc.fit(XTrainLinear, YLinear)
# plotDecisionBoundary(svcR,'SVM using RBF Kernel on Linear Data Set')
                        
# svcR1 = rbf_svc.fit(XTrainParabola, YParabola)
# plotDecisionBoundary(svcR1,'SVM using RBF Kernel on Parabolic Data Set')



```
